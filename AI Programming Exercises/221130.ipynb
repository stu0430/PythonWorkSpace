## 1. 주사위 게임

```python
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import platform
platform.system()
```

```python
if platform.system() == 'Windows':
    plt.rc('font', family='Malgun Gothic')
```

```python
def roll_dice():
    die_1 = random.randint(1, 6)
    die_2 = random.randint(1, 6)
    
    if die_1 == die_2:
        same_num = True
    else:
        same_num = False
        
    return same_num
```

```python
num_simulations = 10000
max_num_rolls = 500
init_balance = 300000
```

```python
win_probality = []
end_balance = []
```

```python
fig = plt.figure()
plt.title('Monte Carlo 주사위 게임')
plt.xlabel('게임 횟수')
plt.ylabel('잔액(원)')
plt.xlim(0, max_num_rolls)
```

### simulation

```python
fig = plt.figure()
plt.title('Monte Carlo 주사위 게임')
plt.xlabel('게임 횟수')
plt.ylabel('잔액(원)')
plt.xlim(0, max_num_rolls)

for i in range(num_simulations):
    balance = [init_balance]
    num_rolls = [0]
    num_wins = 0
    
    for j in range(max_num_rolls):
        same = roll_dice()
        
        if same:
            balance.append(balance[-1] + 4000)
            num_wins += 1
        
        else:
            balance.append(balance[-1] - 1000)
            
            
        num_rolls.append(j + 1)
        
    win_probality.append(num_wins / max_num_rolls)
    end_balance.append(balance[-1])
    
    plt.plot(num_rolls, balance)
    
plt.show()
```
