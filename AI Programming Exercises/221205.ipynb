```python
import random
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rc('font', family='Malgun Gothic')
```

```python
def sim(stock):
    gain = 0
    loss = 1000 * stock
    mean = 557
    std = 183
    buy_prob_instock = 0.05
    buy_prob_outstock = 0.025
    
    guest = int(random.gauss(mean, std))
    
    for g in range(guest):
        if stock > 0:
            if random.random() < buy_prob_instock:
                gain += 10000
                stock -= 1
            
        else:
            if random.random() < buy_prob_outstock:
                gain += 10000
                stock -= 1
                
    return gain - loss
```

```python
num_simulations = 1000
max_stock = 150
```

```python
final_balance = []
```

```python
fig  = plt.figure()
plt.title('재고별 손익')
plt.xlabel('재고(개)')
plt.ylabel('손익(원)')
plt.xlim(0, max_stock)
```

```python
fig = plt.figure()
plt.title('재고별 손익')
plt.xlabel('재고(개)')
plt.ylabel('손익(원)')
plt.xlim(0, max_stock)

final = []

for i in range(num_simulations):
    balance = []
    num_stock = []
    
    for stock in range(0, max_stock):
        balance.append(sim(stock))
        num_stock.append(stock)
        
    plt.plot(num_stock, balance, linewidth=0.1, color='black')
    
    final.append(balance)
    
average = [0] * len(final[0])

for i in range(len(final)):
    for j in range(len(final[0])):
        average[j] += final[i][j] / len(final)
        
plt.plot(num_stock, average, linewidth=3, color='red')
```
