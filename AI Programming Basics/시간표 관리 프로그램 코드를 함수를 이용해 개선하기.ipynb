**시간표 관리 프로그램 코드를 함수를 이용해 개선하기/0527/20220785/이상권**

```python
def relayClass(relay, TimeTable, column, alphabet, period):
    while relay != 1:  # relay가 1이 될 때 까지 연강 시간 만큼 다음 행 동일 열에 알파벳 표시
        TimeTable[period + 3 + (relay - 1)][column] = '|' + alphabet + '|'
        relay -= 1

def Weekend(day):
    print('>>> [ERROR] 토요일과 일요일은 수업이 없습니다!')
    print('>>> 다시 입력하세요.')

def Period_Under_Exceed(period):
    print('>>> [ERROR] 등록하려는 수업이 등록 가능 시간을 벗어납니다!')
    print('>>> 다시 입력하세요.')

def Overlap(period, column):
    print('>>> [ERROR] 해당 시간에 이미 수업이 있습니다!')
    print('>>> 다시 입력하세요.')


# 시간표 형식
TimeTable = [['[             TimeTable              ]'],
             ['--------------------------------------'],
             ['| |', '|M|', '|T|', '|W|', '|T|', '|F|'],
             ['--------------------------------------'],
             ['|1|', '| |', '| |', '| |', '| |', '| |'],  # [4][0] ~ [4][5]
             ['|2|', '| |', '| |', '| |', '| |', '| |'],  # [5][0] ~ [5][5]
             ['|3|', '| |', '| |', '| |', '| |', '| |'],  # [6][0] ~ [6][5]
             ['|4|', '| |', '| |', '| |', '| |', '| |'],  # [7][0] ~ [7][5]
             ['|5|', '| |', '| |', '| |', '| |', '| |'],  # [8][0] ~ [8][5]
             ['|6|', '| |', '| |', '| |', '| |', '| |'],  # [9][0] ~ [9][5]
             ['|7|', '| |', '| |', '| |', '| |', '| |'],  # [10][0] ~ [10][5]
             ['|8|', '| |', '| |', '| |', '| |', '| |'],  # [11][0] ~ [11][5]
             ['--------------------------------------']]

for i in TimeTable:
    print(i)
print()

# 시간표 등록
while True:
    # 요일 및 교시 입력
    day, period = input('>>> 등록할 수업의 요일과 교시를 입력하세요:').split()
    period = int(period)  # 정수 변환
    # 연강 시간 입력
    relay = int(input('>>> 등록할 수업의 연강 시간을 입력하세요:'))
    # 알파벳 입력
    alphabet = input('>>> 등록할 수업의 알파벳을 입력하세요:')

    # 토요일 혹은 일요일은 예외
    if day == '토' or day == '일':
        Weekend(day)
        continue

    # 0교시나 8교시 초과 시 예외
    elif period == 0 or period > 8:
        Period_Under_Exceed(period)
        continue

    else:
        if day == '월':
            column = 1  # 1열
        elif day == '화':
            column = 2  # 2열
        elif day == '수':
            column = 3  # 3열
        elif day == '목':
            column = 4  # 4열
        elif day == '금':
            column = 5  # 5열

        # 시간표 중복 시 예외
        if TimeTable[period + 3][column] != '| |':
            Overlap(period, column)
            continue

        else:  # 알파벳 표시
            TimeTable[period + 3][column] = '|' + alphabet + '|'
            # 연강
            if relay >= 2:  # 2시간 이상 연강
                relayClass(relay, TimeTable, column, alphabet, period)

        for i in TimeTable:
            print(i)

    # 시간표 등록을 계속하거나 종료
    ex_it = int(input('>>> 시간표 등록을 계속하시겠습니까?    1. 예    2. 아니요    '))
    if ex_it == 2:
        print('>>> 프로그램을 종료합니다.')
        break
```
