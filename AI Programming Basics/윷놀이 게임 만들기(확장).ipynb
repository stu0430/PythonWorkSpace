```python
import random

gamers = ('상대방', '당신')
turn = 0
score = [0, 0]

value = ('도', '개', '걸', '윷', '모')

sticks = [0, 0, 0, 0]

end_score = 30

while True:
    print(gamers[turn] + '이 윷을 던졌습니다.')

    count = 0
    for i in range(4):
        sticks[i] = random.randint(0, 1)
        count += sticks[i]

    if count == 0:
        move = 4
    elif count == 1:
        move = 3
    elif count == 2:
        move = 2
    elif count == 3:
        move = 1
    else:
        move = 5

    score[turn] += move

    print('[', end='')
    for i in range(4):
        if sticks[i] == 0:
            print('O', end='')
        else:
            print('X', end='')
    print('] ', end='')

    print(value[move - 1] + '! ' + str(move) + '점을 획득하여 ' +
          gamers[turn] + '의 점수는 현재 ' + str(score[turn]) + '점입니다.')

    if score[turn] >= end_score:
        print(gamers[turn] + '이 이겼습니다!', end='')
        turn = (turn + 1) % 2
        print(gamers[turn] + '의 점수는 현재 ' + str(score[turn]) + '점입니다.')
        break
    if move >= 4:
        print('한 번 더 던질 기회가 주어집니다.')
    else:
        turn = (turn + 1) % 2
        move = 0
        print('다음은 ' + gamers[turn] + '의 차례입니다.')
```
