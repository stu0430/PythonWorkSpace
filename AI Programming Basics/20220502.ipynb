```python
for i in range(1, 101) :
    if i % 3 == 0:
        print('Fizz', end = '')
        if i % 5 == 0 :
            print('Buzz')
        else :
            print()
    elif i % 5 == 0 :
        print('Buzz')
    else :
        print(i)
```

```python
for i in range(1, 101) :
    if i % 3 == 0:
        if i % 5 == 0 :
            print('FizzBuzz')
        else :
            print('Fizz')
    elif i % 5 == 0 :
        print('Buzz')
    else :
        print(i)
```

```python
for i in range(1, 101) :
    if i % 3 == 0 and i % 5 == 0 :
        print('FizzBuzz')
    elif i % 3 == 0 :
        print('Fizz')
    elif i % 5 == 0 :
        print('Buzz')
    else :
        print(i)
```

```python
for i in range(1, 101) :
    print('Fizz' * (i % 3 == 0) + 'Buzz' * (i % 5 == 0) or i)
```
