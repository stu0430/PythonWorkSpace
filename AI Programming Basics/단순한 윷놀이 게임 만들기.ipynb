```python
# 랜덤 모듈
import random

score_me = 0  # 나의 초기 점수
score_opp = 0  # 상대방의 초기 점수

while True:
    while True:  # 상대방의 윷 던지기
        input('윷을 던지려면 Enter를 누르세요.')
        print('\n상대방이 윷을 던졌습니다.')
        roll_opp = random.randint(1, 5)

        if roll_opp == 1:  # 도
            score_opp += 1
            print('도! 1점을 획득하여 상대방의 점수는 현재', str(score_opp) + '점입니다.')
            break

        elif roll_opp == 2:  # 개
            score_opp += 2
            print('개! 2점을 획득하여 상대방의 점수는 현재', str(score_opp) + '점입니다.')
            break

        elif roll_opp == 3:  # 걸
            score_opp += 3
            print('걸! 3점을 획득하여 상대방의 점수는 현재', str(score_opp) + '점입니다.')
            break

        elif roll_opp == 4:  # 윷
            score_opp += 4
            print('윷! 4점을 획득하여 상대방의 점수는 현재', str(score_opp) + '점입니다.')
            print()
            if score_opp < 30:  # 점수가 30점 미만이면 기회 한 번 더
                print('한 번 더 던질 기회가 주어집니다.', '\n')
            elif score_opp >= 30:  # 점수가 30점 이상이면 추가 기회없이 끝내기
                break

        else:  # 모
            score_opp += 5
            print('모! 5점을 획득하여 상대방의 점수는 현재', str(score_opp) + '점입니다.')
            print()
            if score_opp < 30:  # 점수가 30점 미만이면 기회 한 번 더
                print('한 번 더 던질 기회가 주어집니다.', '\n')
            elif score_opp >= 30:  # 점수가 30점 이상이면 추가 기회없이 끝내기
                break

    if score_opp >= 30:  # 점수가 30점 이상일 시 상대 기회로 넘어가지 않고 승자 출력으로
        break

    print('\n다음은 당신 차례입니다.', '\n')

    while True:  # 나의 윷 던지기
        input('윷을 던지려면 Enter를 누르세요.')
        print('\n당신이 윷을 던졌습니다.')
        roll_me = random.randint(1, 5)

        if roll_me == 1:  # 도
            score_me += 1
            print('도! 1점을 획득하여 당신의 점수는 현재', str(score_me) + '점입니다.')
            break

        elif roll_me == 2:  # 개
            score_me += 2
            print('개! 2점을 획득하여 당신의 점수는 현재', str(score_me) + '점입니다.')
            break

        elif roll_me == 3:  # 걸
            score_me += 3
            print('걸! 3점을 획득하여 당신의 점수는 현재', str(score_me) + '점입니다.')
            break

        elif roll_me == 4:  # 윷
            score_me += 4
            print('윷! 4점을 획득하여 당신의 점수는 현재', str(score_me) + '점입니다.')
            print()
            if score_me < 30:  # 점수가 30점 미만이면 기회 한 번 더
                print('한 번 더 던질 기회가 주어집니다.', '\n')
            elif score_me >= 30:  # 점수가 30점 이상이면 추가 기회없이 끝내기
                break

        else:  # 모
            score_me += 5
            print('모! 5점을 획득하여 당신의 점수는 현재', str(score_me) + '점입니다.')
            print()
            if score_me < 30:  # 점수가 30점 미만이면 기회 한 번 더
                print('한 번 더 던질 기회가 주어집니다.', '\n')
            elif score_me >= 30:  # 점수가 30점 이상이면 추가 기회없이 끝내기
                break

    if score_me >= 30:  # 점수가 30점 이상일 시 상대 기회로 넘어가지 않고 승자 출력으로
        break

    print('\n다음은 상대방 차례입니다.', '\n')

# 승자 출력
if score_me >= 30:
    print('당신이 이겼습니다! 상대방의 점수는', str(score_opp) + '점입니다.')

elif score_opp >= 30:
    print('상대방이 이겼습니다! 당신의 점수는 ' + str(score_me) + '점입니다.')
```
