**시간표 관리 프로그램을 파일을 이용해 확장하기/0604/20220785/이상권**

```python
import pickle
import os

def relayClass(relay, TimeTable, day, alphabet, period):
    if relay >= 2:
        while relay != 1:  # relay가 1이 될 때 까지 연강 시간 만큼 다음 행 동일 열에 알파벳 표시
            TimeTable[str(int(period) + (relay - 1))][day] = alphabet
            relay -= 1

def Weekend(day):
    print('>>> [ERROR] 토요일과 일요일은 수업이 없습니다!')
    print('>>> 다시 입력하세요.')

def Period_Under_Exceed(period):
    print('>>> [ERROR] 등록하려는 수업이 등록 가능 시간을 벗어납니다!')
    print('>>> 다시 입력하세요.')

def Overlap(period, day):
    print('>>> [ERROR] 해당 시간에 이미 수업이 있습니다!')
    print('>>> 다시 입력하세요.')

def print_TimeTable(**classdict):
    print('\n')
    print('[ TimeTable ]')
    print('-------------')
    print('| |M|T|W|T|F|')               
    print('-------------')

    for hour in classdict.keys():
        print('|', end = '')
        print(hour, end = '|')         
        for day in classdict[hour].keys():
            print(classdict[hour][day], end = '|')  
        print()

    print('-------------\n')

def saveTimeTabletxt(**classdict):
    f = open('timetable.txt', 'w')
    print('\n', file = f)
    print('[ TimeTable ]', file = f)
    print('-------------', file = f)
    print('| |M|T|W|T|F|', file = f)
    print('-------------', file = f)

    for hour in classdict.keys():
        print('|', end = '', file = f)
        print(hour, end = '|', file = f)
        for day in classdict[hour].keys():
            print(classdict[hour][day], end = '|', file = f)
        print(file = f)

    print('-------------\n', file = f)
    f.close()

weekdays = ['월', '화', '수', '목', '금']
hours = [str(x) for x in range(1, 9)]

TimeTable = {key: {key: ' ' for key, value in dict.fromkeys(weekdays).items()} for key in dict.fromkeys(hours).keys()}

print_TimeTable(**TimeTable)

# 시간표 등록
while True:
    # 요일 및 교시 입력
    day, period = input('>>> 등록할 수업의 요일과 교시를 입력하세요:').split()
    # 연강 시간 입력
    relay = int(input('>>> 등록할 수업의 연강 시간을 입력하세요:'))
    # 알파벳 입력
    alphabet = input('>>> 등록할 수업의 알파벳을 입력하세요:')

    # 토요일 혹은 일요일은 예외
    if day == '토' or day == '일':
        Weekend(day)
        continue

    # 0교시나 8교시 초과 시 예외
    elif int(period) <= 0 or int(period) > 8:
        Period_Under_Exceed(period)
        continue

    else:
        # 시간표 중복 시 예외
        if TimeTable[period][day] != ' ':
            Overlap(period, day)
            continue

        else:  # 알파벳 표시
            TimeTable[period][day] = alphabet
            # 연강
            relayClass(relay, TimeTable, day, alphabet, period)

    print_TimeTable(**TimeTable)

    # 시간표 등록을 계속하거나 종료
    ex_it = int(input('>>> 시간표 등록을 계속하시겠습니까?    1. 예    2. 아니요    '))
    if ex_it == 2:
        print('>>> 시간표 등록을 종료합니다.')
        break

save = int(input('>>> 등록된 시간표를 저장하시겠습니까?    1. 예    2.아니요    '))
if save == 1:
    save_sel = int(input('>>> 저장할 방식을 선택하세요:    1. 텍스트 저장    2. 딕셔너리 저장    3. 둘 다 저장 선택하기    '))
    if save_sel == 1:
        saveTimeTabletxt(**TimeTable)
        print('>>> 텍스트로 저장되었습니다.')

    elif save_sel == 2:
        with open('timetable.pkl', 'wb') as f:
            pickle.dump(TimeTable, f)
        print('>>> 딕셔너리로 저장되었습니다.')

    elif save_sel == 3:
        saveTimeTabletxt(**TimeTable)
        with open('timetable.pkl', 'wb') as f:
            pickle.dump(TimeTable, f)
        print('>>> 텍스트와 딕셔너리 둘 다 저장되었습니다.')


load = int(input('>>> 저장된 시간표를 불러오시겠습니까?    1. 예    2.아니요    '))
if load == 1:
    load_sel = int(input('>>> 불러올 파일을 선택하세요:    1. 텍스트 불러오기    2. 딕셔너리 불러오기    '))
    if load_sel == 1:
        if(os.path.isfile('timetable.txt')):
            with open('timetable.txt', 'r') as file:
                for text in file:
                    print(text.strip('\n'))
        else:
            print('>>> [ERROR] 불러올 시간표가 없습니다!')

    elif load_sel == 2:
        if (os.path.isfile('timetable.pkl')):
            with open('timetable.pkl', 'rb') as f:
                mydict = pickle.load(f)
                print_TimeTable(**mydict)
        else:
            print('>>> [ERROR] 불러올 시간표가 없습니다!')
```
